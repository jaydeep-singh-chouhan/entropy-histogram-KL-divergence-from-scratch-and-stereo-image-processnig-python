```python
def func(x,i):
  if x>0 and x<=(1/3) :
    return 1
  elif x>(1/3) and x<=(1/3)*(1+(1/i)):
    return 1-(3*i*x-(i))
  elif x>(1/3)*(1+(1/i)) and x<=1:
    return 0
```

**Entropy function**

```python
def entropy(pmf):
  x=0
  for i in range(len(pmf)):
    if(pmf[i]==0): # checking for the zero if pmf==0 then skip 
      x=x+0
    else:
      x=x+pmf[i]*math.log(pmf[i],2)
  return -x
```

```python
entropy(dist)
```

```python
entropy_arr=np.zeros(20) #creating an empty entropy array
pmf_values=np.arange(0,1,0.05) #generating pmf values
for i in range(20):
  p=pmf_values[i]
  pmf=np.array([p,1-p])
  entropy_arr[i]=entropy(pmf)
```

```python
plt.grid()
plt.plot(pmf_values,entropy_arr,'g-')
plt.show()
```

**Normal histogram function for 2D image**

```python
def joint_hist(p,q):
  row,column=p.shape
  #x=np.zeros((row*column),2)
  hist=np.zeros((256,256))
  #k=0
  for i in range(row):
    for j in range(column):
      #x[k]=[p[i][j],q[i][j]]
      m =p[i][j]
      n=q[i][j]
      hist[m][n]+=1
      #k=k+1
  normal_hist=hist/(row*column)
  return normal_hist
```

```python
#here index i am assuming p(x/y) then for y it is 1 and if it is p(y/x) then it is 2
def conditional_pmf_func(j_pmf,index,value):
   
    rows,columns=j_pmf.shape
    marginal=0
    if(index==1):
      conditonal_pmf=np.zeros(columns)
      for i in range(columns):
        marginal=marginal+j_pmf[value][i]
      for i in range(columns):
        conditonal_pmf[i]=j_pmf[value][i]/marginal
    else:
      conditonal_pmf=np.zeros(rows)
      for i in range(rows):
        marginal=marginal+j_pmf[i][value]
      for i in range(columns):
        conditonal_pmf[i]=j_pmf[i][value]/marginal

    return conditonal_pmf
```

**B) CONDITIONAL ENTROPY**

```python
#if index =1 the H(x/y) if index=2 then H(y/x)
def conditional_entropy(j_pmf,index):
  rows,columns=j_pmf.shape
  if index==1:
    H_conditional=0
    for i in range(rows):
      for j in range(columns):
        conditional=func(j_pmf,index,i)
        if(conditional[j]==0):
          continue
        H_conditional=H_conditional+j_pmf[i][j]*math.log(conditional[j],2)
  else:
    H_conditional=0
    for i in range(columns):
      for j in range(rows):
        conditional=func(j_pmf,index,i)
        if(conditional[j]==0):
          continue
        H_conditional=H_conditional+j_pmf[i][j]*math.log(conditional[j],2)
  return -H_conditional  
```

**Q7 KL divergence**

```python
def Kl_divergence(p,q):
  D=0
  for i in range(len(p)):
    if q[i]==0:
      continue
    x=p[i]/q[i]
    
    D+=p[i]*math.log(x,2)
  return D
```

```python
# generating binomial distribution for testing my function 
from scipy.stats import binom
import numpy as np
import matplotlib.pyplot as plt
import math

# setting the values
# of n and p
n = 6
p = 0.6
n1=6
p1=0.4
# defining list of r values
r_values = list(range(n + 1))
# list of pmf values
dist1 = np.array([binom.pmf(r, n, p) for r in r_values ])
dist2 =np.array([binom.pmf(r, n1, p1) for r in r_values ])
```
